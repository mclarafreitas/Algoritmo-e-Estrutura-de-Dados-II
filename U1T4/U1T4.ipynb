```python
import osmnx as ox
import networkx as nx

# Baixar a rede de ruas de Natal, RN
G = ox.graph_from_place("Natal, Rio Grande do Norte, Brazil", network_type="drive")

# Mostrar o grafo
fig, ax = ox.plot_graph(G, node_size=5, node_color='blue', edge_linewidth=0.5)
```

```python
# Extrair o maior componente fortemente conectado
G_largest = max(nx.strongly_connected_components(G), key=len)
G_largest = G.subgraph(G_largest).copy()

# Converter o grafo para não-direcionado e simples
G_simple = nx.Graph(G_largest)

# Calcular as métricas no maior componente conectado e simplificado
avg_shortest_path_length = nx.average_shortest_path_length(G_simple)
diameter = nx.diameter(G_simple)
clustering_coefficient = nx.average_clustering(G_simple)

# Exibir as métricas
print("Average Shortest Path Length:", avg_shortest_path_length)
print("Diameter of Network:", diameter)
print("Clustering Coefficient:", clustering_coefficient)
```
