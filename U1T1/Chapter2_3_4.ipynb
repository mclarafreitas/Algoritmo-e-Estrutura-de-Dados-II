2. 60% of the emails hitting my inbox is spam. You design a phenomenal spam filter which is able to tell me, with 98% accuracy, whether an email is spam or not: if an email is not spam, the system has a 98% probability of saying so. The filter knows 60% of emails are spam and so it will flag 60% of my emails. Suppose that, at the end of the week, I look in my spam box and see 963 emails. Use Bayes’ Theorem to calculate how many of those 963 emails in my spam box I should suspect to be non-spam.

According to Michele Coscia in the book The Atlas For The Aspiring Network Scientist, Bayes’ Theorem is an almost magical formula that allows you to
 estimate the probability of an event based on your priors.
 Keeping the example of cheating on a coin toss, we want to estimate the
 probability I cheated and rigged the coin so it lands on heads after
 we tossed it and it indeed landed on heads– in mathematical notation: P(W|H). To do so, you need to have priors. You need to know: what’s the probability of heads for all coins in the world (whether they are rigged or not, P(H)), what’s the probability I rigged the coin (P(W)), and what is the probability of obtaining heads on a rigged coin (P(H|W)). Without further ado, here’s one of the most important formulas in human history:

image.png
Coscia, Michele. The Atlas For The Aspiring Network Scientist.2021.

This is a classic problem involving Bayes’ Theorem, which can help us calculate the probability that an email flagged as spam by your filter is actually non-spam (also known as a false positive). Lets think in the setup of the problem:

* Let (P(Spam) = 0.60) be the probability that an email is spam.
* ( P(Not Spam) = 1 - P(Spam) = 0.40 ) is the probability that an email is not spam.
* The filter has a 98% accuracy in correctly identifying spam and non-spam emails:
  * ( P(Correct|Spam) = 0.98), the probability that the filter correctly identifies spam.
  * ( P(Correct|Not Spam) = 0.98), the probability that the filter correctly identifies non-spam.
  * This implies that the false positive rate is ( P(Spam|Not Spam) = 1 - 0.98 = 0.02)

We are asked to determine how many of the 963 flagged emails should be suspected to be non-spam. In other words, we want to calculate (P(Not Spam| Flagged as Spam)), which tells us the probability that an email flagged as spam is actually non-spam.

Lets Apply Bayes' Theorem
P(Not Spam | Flagged as Spam) = P(Flagged as Spam | Not Spam) * P(Not Spam) / P(Flagged as Spam)

We need to find P(Flagged as Spam), because it will say the total probability that an email is flagged as spam. We can use the law of total probability to compute this:
P(Flagged as spam) = P(Flagged as spam|spam) * P(spam) + P(Flagged as spam|Not Spam) * P(Not spam)
Where:
* P(Spam) = 0.60
* P(Flagged as spam|Spam) = 0.98
* P(Flagged as spam|Not spam) = 0.02

So now, lets compute the probabilities:
P(Flagged as spam) = (0.98 * 0.60) + (0.02 * 0.40) = 0.596

And now, we apply the Bayes Theorem:
P(Not Spam|Flagged as spam) = 0.02 * 0.40 / 0.596 ˜= 0.0134

The probability that an email flagged as spam is actually non-spam is approximately 0.0134 or 1.34%

And finally, lets calculate the number of non-spam emails.
If you have 963 emails flagged as spam, the number of emails that are non-spam is:
Non-spam emails = 0.0134 * 963 ˜= 12.93

12.93 emails out of 963 flagged emails are likely to be non-spam.

4. Draw the probability mass function and the cumulative distribution of the following outcome probabilities:

image.png

Probability Mass Function (PMF) show us the probability assined to each outcome.

Cumulative Distribution Function (CDF) show us the cumulative sum of probabilities as we move through the outcome.

To solve this problem I will make a python code that solves the problem and plot a graphic.

```python
import matplotlib.pyplot as plt
import numpy as np

# Given outcome probabilities
outcomes = np.array([1, 2, 3, 4, 5, 6, 7, 8])
probabilities = np.array([0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02])

# Cumulative probabilities
cumulative_probabilities = np.cumsum(probabilities)

# Plotting the PMF and CDF
fig, ax = plt.subplots(2, 1, figsize=(8, 10))

# Plot PMF
ax[0].bar(outcomes, probabilities, color='red', edgecolor='black')
ax[0].set_title('Probability Mass Function (PMF)')
ax[0].set_xlabel('Outcome')
ax[0].set_ylabel('Probability')
ax[0].set_xticks(outcomes)

# Plot CDF
ax[1].step(outcomes, cumulative_probabilities, where='mid', color='red', linewidth=2)
ax[1].set_title('Cumulative Distribution Function (CDF)')
ax[1].set_xlabel('Outcome')
ax[1].set_ylabel('Cumulative Probability')
ax[1].set_xticks(outcomes)

# Display the plots
plt.tight_layout()
plt.show()

# Display numerical values for PMF and CDF
for i in range(len(outcomes)):
    print(f"Outcome {outcomes[i]}: P(X = {outcomes[i]}) = {probabilities[i]}, CDF = {cumulative_probabilities[i]}")
```

Here are the two graphs based on the provided probabilities:

Probability Mass Function (PMF): This bar chart shows the probability for each outcome.

Cumulative Distribution Function (CDF): This step plot shows the cumulative probabilities as you move through the outcomes.

Explanation of the code:
 * PMF Calculation: The array probabilities stores the probability for each outcome. This is plotted using a bar chart.
 * CDF Calculation: The np.cumsum() function is used to compute the cumulative sum of probabilities, which gives the CDF.
 * Plotting: The bar chart shows the PMF, and the step plot shows the CDF. The results are printed for each outcome.

 5. How many bits do we need to independently encode v1 and v2 fromhttp://www.networkatlas.eu/exercises/2/5/data.txt?
How much would we save in encoding v1 if we knew v2?

To solve this problem we need to calculate the entropy of the distribuition in v1 and v2.

Let's start:
* 1. We need to compute the entropy of v1 and v2. Entropy is a measure of uncertainty or randomness in a variable. It quantifies how much information is needed to represent a random variable. In this context, the more uncertain or random a variable is, the more bits it takes to encode it.

The formula of entropy is: H(X)=−∑p(x)log2p(x)

* 2. Mutual Information (MI) measures how much information one variable (say v2) gives us about another variable (v1). It quantifies the reduction in uncertainty about v1 that results from knowing v2.
In terms of encoding, MI helps us calculate how much we can reduce the number of bits required to encode v1 if we have knowledge of v2.
The formula for mutual information between two variables X and Y is:

𝐼(𝑋;𝑌)=𝐻(𝑋)−𝐻(𝑋∣𝑌)I(X;Y)=H(X)−H(X∣Y)

* The entropy of v1 tells us how many bits are needed to encode v1 if we know nothing else.
The mutual information tells us how many bits can be saved if we know v2 when encoding v1.

```python
pip install pandas
```

```python
import pandas as pd
from scipy.stats import entropy
from sklearn.metrics import mutual_info_score

# Create the DataFrame directly from the provided data
data = {
    "v1": ['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'b', 'b', 'b', 'b', 'b', 'b', 'c', 'c', 'c', 'c', 'c', 'c',
           'c', 'd', 'd', 'd', 'd', 'd', 'd', 'd', 'd', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e',
           'f', 'f', 'f', 'f', 'f'],
    "v2": ['a', 's', 'a', 'd', 'd', 'f', 'f', 'f', 'f', 'g', 'g', 'g', 'g', 'h', 'h', 'h', 'h', 'j', 'j', 'j',
           'k', 'j', 'g', 'h', 'h', 'g', 'f', 'f', 'f', 'f', 'd', 'd', 's', 's', 's', 's', 's', 'a', 'a', 'a',
           'a', 'a', 'd', 'f', 'g']
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Calculate the probabilities of each unique value in v1 and v2
v1_ps = (df.groupby(by="v1").size() / df.shape[0]).values
v2_ps = (df.groupby(by="v2").size() / df.shape[0]).values

# Calculate the entropy for v1 and v2
v1_info = entropy(v1_ps)
v2_info = entropy(v2_ps)

# Print the entropies (amount of information needed to encode each)
print(f"Entropy of v1: {v1_info:.2f} bits")
print(f"Entropy of v2: {v2_info:.2f} bits")

# Calculate the mutual information between v1 and v2 (bits saved by knowing v2 when encoding v1)
mi = mutual_info_score(df["v1"], df["v2"])

# Calculate and print the percentage of bits saved
percentage_saved = 100 * mi / v1_info
print(f"Knowing v2 gives us {percentage_saved:.2f}% of v1's information")
```

The code uses entropy to measure how many bits are needed to encode v1 and v2 independently, and then it uses mutual information to calculate how much information we can save when encoding v1 if we already know v2. This is useful for data compression, as reducing redundancy between variables allows us to store or transmit data more efficiently.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Cria um grafo não direcionado
G = nx.Graph()

# Adiciona nós (pessoas no grafo)
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
G.add_nodes_from(nodes)

# Adiciona arestas ponderadas baseadas nas descrições
edges = [
    ('A', 'B', 1),   # A considera B amigo, não é recíproco
    ('B', 'C', 2),   # B e C têm amizade recíproca
    ('B', 'D', 2),   # B e D têm amizade recíproca
    ('C', 'D', 1),   # C considera D amigo, não é recíproco
    ('D', 'E', 1),   # D enviou solicitação para E, não é recíproco
    ('D', 'F', 1),   # D enviou solicitação para F, não é recíproco
    ('D', 'G', 1),   # D enviou solicitação para G, não é recíproco
    ('D', 'H', 1),   # D enviou solicitação para H, não é recíproco
    ('G', 'D', 2),   # G respondeu para D, então a relação é recíproca
    ('A', 'G', 2)    # A e G têm amizade recíproca
]

# Adiciona as arestas e pesos no grafo
G.add_weighted_edges_from(edges)

# Desenha o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Layout do grafo

# Desenha os nós e arestas
nx.draw(G, pos, with_labels=True, node_color='lightblue', font_weight='bold', node_size=2000)

# Desenha os pesos das arestas
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.title("Grafo Não Direcionado e Ponderado das Amizades")
plt.show()
```

4. Draw a correlation network for the vectors in http://www.networkatlas.
 eu/exercises/3/4/data.txt, by only drawing edges with positive
 weights, ignoring self loops.




We need to:

* Compute the correlation between each pair of vectors (columns) in the matrix.
* Ignore negative correlations and self-loops (where a node connects to itself).
* Draw a network where only positive correlations are represented as edges between nodes (vectors).

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Defina a matriz diretamente no código
data = [
    [1, 6, 1, 1, 1, 3, 1, 1],
    [1, 4, 4, 6, 1, 5, 3, 7],
    [1, 4, 4, 6, 1, 5, 7, 8],
    [1, 4, 6, 8, 1, 7, 8, 4],
    [1, 5, 7, 7, 2, 0, 0, 5],
    [2, 2, 5, 7, 2, 9, 9, 5],
    [3, 6, 7, 0, 3, 7, 6, 4],
    [4, 6, 9, 5, 3, 3, 3, 7],
    [4, 4, 8, 4, 4, 2, 3, 5],
    [3, 6, 7, 3, 5, 0, 3, 5],
    [5, 4, 7, 2, 5, 8, 7, 5],
    [6, 6, 9, 4, 6, 8, 7, 4],
    [7, 4, 7, 5, 7, 4, 5, 7],
    [8, 6, 0, 2, 7, 2, 4, 3],
    [9, 7, 3, 9, 8, 3, 5, 6]
]

# Converta a matriz em um DataFrame do Pandas
columns = ["A", "B", "C", "D", "E", "F", "G", "H"]
df = pd.DataFrame(data, columns=columns)

# Calcule a correlação
corr_matrix = df.corr()

# Considera apenas as correlações positivas
df_pos_corr = corr_matrix[corr_matrix > 0].unstack().reset_index()
df_pos_corr.columns = ["source", "target", "edge"]
df_pos_corr = df_pos_corr[df_pos_corr["edge"] > 0]

# Remova os loops (mesmas fontes e alvos)
df_pos_corr = df_pos_corr[df_pos_corr["source"] != df_pos_corr["target"]]

# Crie o objeto de grafo
G = nx.from_pandas_edgelist(df_pos_corr, "source", "target", ["edge"])

# Desenhe o grafo
nx.draw(G, with_labels=True, node_size=700, node_color="red", font_size=10, font_weight="bold")
plt.show()

```

Code Explanation:

* Compute Correlations: The code calculates the Pearson correlation between all pairs of vectors.
* Filter Positive Correlations: It keeps only positive correlations and removes negative or zero correlations, as well as self-loops (i.e., where a node is connected to itself).
* Create a Network: A graph is built using the positive correlations, where each node is a vector, and an edge exists if there is a positive correlation between two vectors.
* Visualize the Graph: The resulting correlation network is drawn, showing the relationships between the vectors based on their positive correlations.

This process turns a matrix of data into a network, highlighting how the different vectors relate to each other based on positive correlation.

The edge type is either the numerical id of the layer, or “C” for an inter-layer coupling. Given that this is a one-to-one multilayer network, determine whether this network has a star, clique or chain coupling.

```python
from collections import defaultdict

# Define the edges as given in the data
data = [
    (1, 2, 'C'), (1, 3, 'C'), (1, 4, 'C'),
    (5, 6, 'C'), (5, 7, 'C'), (5, 8, 'C'),
    (9, 10, 'C'), (9, 11, 'C'), (9, 12, 'C'),
    (13, 14, 'C'), (13, 15, 'C'), (13, 16, 'C'),
    (1, 5, 1), (1, 9, 1), (1, 13, 1),
    (5, 9, 1), (5, 13, 1), (9, 13, 1),
    (2, 6, 2), (6, 10, 2), (10, 14, 2),
    (14, 2, 2), (3, 7, 3), (7, 11, 3),
    (11, 15, 3), (4, 8, 4), (12, 16, 4)
]

# Create adjacency lists for each layer
layer_dict = defaultdict(list)

for source, target, edge_type in data:
    layer_dict[edge_type].append((source, target))

# Check for star, clique, or chain coupling
def check_coupling(layer_dict):
    star_found = False
    clique_found = True
    chain_found = True

    for layer, edges in layer_dict.items():
        if layer == 'C':
            continue  # Skip inter-layer couplings for clique and chain check

        # Build adjacency list for this layer
        layer_graph = defaultdict(set)
        for source, target in edges:
            layer_graph[source].add(target)
            layer_graph[target].add(source)

        # Check for star structure
        center_candidates = [node for node, neighbors in layer_graph.items() if len(neighbors) > 1]
        if center_candidates:
            star_found = True

        # Check for clique structure
        total_nodes = len(layer_graph)
        edge_count = sum(len(neighbors) for neighbors in layer_graph.values()) // 2
        if edge_count != total_nodes * (total_nodes - 1) // 2:
            clique_found = False

        # Check for chain structure
        if any(len(neighbors) != 2 for node, neighbors in layer_graph.items() if len(neighbors) > 0):
            chain_found = False

    return star_found, clique_found, chain_found

star, clique, chain = check_coupling(layer_dict)

# Output results
if star:
    print("The network has a star coupling.")
else:
    print("No star coupling detected.")

if clique:
    print("The network has a clique coupling.")
else:
    print("No clique coupling detected.")

if chain:
    print("The network has a chain coupling.")
else:
    print("No chain coupling detected.")
```

Explanation of the code:

* Input Data: The data contains edges between nodes, where each edge has a source, a target, and an edge type. The edge type can either be a layer ID (1, 2, 3, 4) or "C" (representing inter-layer couplings).

* Organize Edges by Layer: The edges are grouped into layers using a dictionary layer_dict, where each layer ID is associated with the edges in that layer.

* Check Coupling Types:
   - Star Structure: The code checks if any node has multiple connections (more than one neighbor) in a layer, which would suggest a star structure. If such a node is found, the star structure is flagged as found.
   - Clique Structure: A clique means every node in the layer is connected to every other node. The code checks if the number of edges matches the number required for a complete graph, which would indicate a clique.
   - Chain Structure: In a chain, each node is connected to exactly two neighbors (except for the endpoints, which have one neighbor). The code checks if all nodes (except endpoints) have exactly two neighbors.

* Determine Results: The code determines whether the network contains a star, clique, or chain coupling based on the checks.

* Output: It prints whether each type of coupling (star, clique, chain) was detected.

So, the code analyses the multilayer network to detect if the structure follows a star, clique, or chain pattern, by examining the connections in each layer separately and comparing them against the criteria for each coupling type.

3. The network in http://www.networkatlas.eu/exercises/4/3/data.txt is a hypergraph, with a hyperedge per line. Transform it in a unipartite network in which each hyperedge is split in edges connecting all nodes in the hyperedge. Then transform it into a bipartite network in which each hyperedge is a node of one type and its nodes connect to it.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the hypergraph edges as given in the data
hyperedges = [
    [1, 2, 3, 4, 5, 6],
    [2, 4, 7, 8],
    [4, 5, 6, 7, 10],
    [6, 9, 10, 11, 12],
    [11, 13, 14, 15],
    [11, 12, 13],
    [14, 15, 16, 17, 18],
    [15, 17, 19, 20],
    [21, 22, 23],
    [23, 24],
    [24, 25, 26],
    [25, 27, 28, 29]
]

# Step 1: Create Unipartite Network
unipartite_graph = nx.Graph()

# Adding edges between all pairs in each hyperedge
for hyperedge in hyperedges:
    unipartite_graph.add_edges_from([(node1, node2) for i, node1 in enumerate(hyperedge) for node2 in hyperedge[i+1:]])

# Draw Unipartite Network
plt.figure(figsize=(10, 6))
nx.draw(unipartite_graph, with_labels=True, node_color='lightblue', edge_color='gray', node_size=700)
plt.title("Unipartite Network from Hypergraph")
plt.show()

# Step 2: Create Bipartite Network
bipartite_graph = nx.Graph()

# Adding nodes for hyperedges and connecting them to their nodes
for idx, hyperedge in enumerate(hyperedges):
    hyperedge_node = f"Hyperedge_{idx+1}"
    bipartite_graph.add_node(hyperedge_node, bipartite=0)  # Type 0 for hyperedges
    for node in hyperedge:
        bipartite_graph.add_node(node, bipartite=1)  # Type 1 for nodes
        bipartite_graph.add_edge(hyperedge_node, node)

# Draw Bipartite Network
plt.figure(figsize=(10, 6))
pos = nx.bipartite_layout(bipartite_graph, nodes=[f"Hyperedge_{i+1}" for i in range(len(hyperedges))])
nx.draw(bipartite_graph, pos, with_labels=True, node_color=['lightgreen' if bipartite_graph.nodes[node]['bipartite'] == 0 else 'lightcoral' for node in bipartite_graph.nodes], edge_color='gray', node_size=700)
plt.title("Bipartite Network from Hypergraph")
plt.show()
```

Explanation of the code:

* Input Data: The code starts with a list of hyperedges, where each hyperedge is represented as a list of nodes.

* Create a Unipartite Network:
   - A new unipartite graph is created using the NetworkX library.
   - For each hyperedge, edges are created between all pairs of nodes within that hyperedge. This is done using a list comprehension that generates combinations of nodes, effectively connecting every node in the hyperedge to every other node.
   - The resulting unipartite graph represents all the connections among the nodes as defined by the hyperedges.

* Visualization of the Unipartite Network: The unipartite graph is drawn and displayed, showing how nodes are interconnected.

* Create a Bipartite Network:
   - A new bipartite graph is created.
   - Each hyperedge is represented as a unique node in one layer (type 0), while the nodes in the hyperedges are added as nodes in another layer (type 1).
   - Edges are created between each hyperedge node and its corresponding nodes, establishing connections that reflect the original hyperedges.

* Visualization of the Bipartite Network: The bipartite graph is drawn and displayed, showing the hyperedges and their connections to the individual nodes.

So, the code transforms the hypergraph into a unipartite network by connecting nodes within each hyperedge, and then into a bipartite network where hyperedges are treated as nodes that connect to their constituent nodes, effectively illustrating the relationships defined by the hyperedges.

4. The network in http://www.networkatlas.eu/exercises/4/4/data.txt is dynamic, the third and fourth columns of the edge list tell you the first and last snapshot in which the edge was continuously present. An edge can reappear if the edge was present in two discontinuous time periods. Aggregate it using a
disjoint window of size 3.


```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from io import StringIO

# Load the data as a string
data = """
1	2	1	4
1	2	7	9
1	3	3	7
1	4	2	4
1	5	1	9
2	3	1	2
2	4	1	3
2	4	6	7
2	4	9	9
3	5	5	7
3	6	1	3
3	6	8	9
4	7	1	9
5	8	1	2
5	8	4	7
5	8	9	9
5	9	1	5
5	10	8	9
5	11	7	8
6	12	4	7
6	13	2	3
7	12	2	6
"""

# Parse the data into a MultiGraph
G = nx.read_edgelist(StringIO(data), create_using=nx.MultiGraph(), data=[("start", int), ("end", int)])

# Create graphs for each time window
first_window = nx.Graph()
second_window = nx.Graph()
third_window = nx.Graph()

# Add edges to respective graphs based on their time ranges
for e in G.edges(data=True):
    if e[2]["start"] <= 3:
        first_window.add_edge(e[0], e[1])
    if e[2]["start"] <= 6 and e[2]["end"] > 3:
        second_window.add_edge(e[0], e[1])
    if e[2]["end"] > 6:
        third_window.add_edge(e[0], e[1])

# Draw the graphs for visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
nx.draw(first_window, with_labels=True, node_color='skyblue', node_size=700, font_size=10)
plt.title('Time Window 1 (1-3)')

plt.subplot(1, 3, 2)
nx.draw(second_window, with_labels=True, node_color='lightgreen', node_size=700, font_size=10)
plt.title('Time Window 2 (4-6)')

plt.subplot(1, 3, 3)
nx.draw(third_window, with_labels=True, node_color='salmon', node_size=700, font_size=10)
plt.title('Time Window 3 (7-9)')

plt.show()

```

Explanation of the code:

* Input Data: The code starts with a string representation of a dynamic network, where each line represents an edge between two nodes, along with the time periods (start and end) during which the edge is active.

* Load Data into a MultiGraph:
   - The data is parsed into a MultiGraph using the NetworkX library. This allows multiple edges between the same pair of nodes, which is useful for dynamic networks.

* Create Graphs for Each Time Window:
   - Three separate graphs are created to represent different disjoint time windows (1-3, 4-6, and 7-9).
   - The code iterates through the edges in the original graph:
     - First Window: An edge is added to the first window graph if its start time is less than or equal to 3.
     - Second Window: An edge is added to the second window graph if its start time is less than or equal to 6 and its end time is greater than 3, indicating the edge was active during part of this window.
     - Third Window: An edge is added to the third window graph if its end time is greater than 6.

* Visualization of the Time Windows:
   - The three graphs representing the different time windows are drawn and displayed using Matplotlib.
   - Each subplot corresponds to one time window, with distinct colors for better visualization.

* Output: The resulting visualizations show the connectivity of the network across the three time windows, allowing analysis of how the network evolves over time.

So, the code effectively aggregates the dynamic network into three disjoint time windows, representing how edges appear based on their active time periods, and visualizes each window to illustrate the state of the network at different times.
